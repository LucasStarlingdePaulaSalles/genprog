```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tour_p50_g50_09_005 = pd.read_csv('exp/tour-p50-g50-09-005.csv')
tour_p50_g100_09_005= pd.read_csv('exp/tour-p50-g100-09-005.csv')
tour_p50_g500_09_005= pd.read_csv('exp/tour-p50-g500-09-005.csv')
tour_p100_g50_09_005= pd.read_csv('exp/tour-p100-g50-09-005.csv')
tour_p100_g100_09_005= pd.read_csv('exp/tour-p100-g100-09-005.csv')
tour_p100_g500_09_005= pd.read_csv('exp/tour-p100-g500-09-005.csv')
tour_p500_g50_09_005= pd.read_csv('exp/tour-p500-g50-09-005.csv')
tour_p500_g100_09_005= pd.read_csv('exp/tour-p500-g100-09-005.csv')
tour_p500_g500_09_005= pd.read_csv('exp/tour-p500-g500-09-005.csv')
```

```python
tour_p50_g50_09_005.head()
```

```python
min_tour_by_gen = pd.DataFrame({
    "Pop: 50 & Gen: 50": tour_p50_g50_09_005["min"],
    "Pop: 50 & Gen: 100": tour_p50_g100_09_005["min"],
    "Pop: 50 & Gen: 500": tour_p50_g500_09_005["min"],
    "Pop: 100 & Gen: 50": tour_p100_g50_09_005["min"],
    "Pop: 100 & Gen: 100": tour_p100_g100_09_005["min"],
    "Pop: 100 & Gen: 500": tour_p100_g500_09_005["min"],
    "Pop: 500 & Gen: 50": tour_p500_g50_09_005["min"],
    "Pop: 500 & Gen: 100": tour_p500_g100_09_005["min"],
    "Pop: 500 & Gen: 500": tour_p500_g500_09_005["min"],
  })

sns.set(rc={'figure.figsize':(15, 10)})
sns.lineplot(data=min_tour_by_gen[[
    "Pop: 50 & Gen: 50", "Pop: 50 & Gen: 50", "Pop: 50 & Gen: 500",
    "Pop: 100 & Gen: 50", "Pop: 100 & Gen: 100", "Pop: 100 & Gen: 500",
    "Pop: 500 & Gen: 50", "Pop: 500 & Gen: 100", "Pop: 500 & Gen: 500",
    ]])
plt.xlabel("Gerações")
plt.ylabel("Fitness")
plt.title("Fitness dos melhores indivíduos por geração")
plt.xlim([0, 500])
# plt.ylim([6.0, 7.0])
plt.show()
```
